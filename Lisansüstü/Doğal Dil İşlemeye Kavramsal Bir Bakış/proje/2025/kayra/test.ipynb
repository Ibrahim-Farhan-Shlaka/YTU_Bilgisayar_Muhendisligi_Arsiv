```python
model_save_path = os.path.join(kok_dizin, "egitilen_model")
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Cihaz:{device}")
```

# Tahmin etme fonksiyonu ekle

```python
# Tahmin fonksiyonu
def tahmin_yap(metin, model, tokenizer, device):
    inputs = tokenizer(metin, truncation=True, padding=True, return_tensors="pt").to(device)
    outputs = model(**inputs)
    probs = outputs.logits.softmax(dim=-1)
    prediction = torch.argmax(probs, dim=-1).item()
    confidence = probs[0][prediction].item()

    # Etiketler (0: normal, 1: spam olarak varsayıyorum, modelinize göre değiştirin)
    labels = ["Spam", "Normal"]

    return {
        "metin": metin,
        "tahmin": labels[prediction],
        "güven": f"{confidence:.4f}"
    }
```

# Modeli oku

```python
# Model'i yükle
model = AutoModelForSequenceClassification.from_pretrained(model_save_path)
model.to(device)
# Tokenizer'ı yükle
tokenizer = AutoTokenizer.from_pretrained(model_save_path)
```

# Girdiyi tahminle

```python
# Örnek metin 'Tamam uyar yarn gidelim;)'
while True:
    metin = input("Metni girin (çıkmak için 'q' yazın): ")
    if metin.lower() == 'q':
        break

    # Tahmin yap
    sonuc = tahmin_yap(metin, model, tokenizer, device)

    # Sonucu yazdır
    print(f"Metin: {sonuc['metin']}\n Tahmin: {sonuc['tahmin']}, Güven: {sonuc['güven']}")
```
