```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import os
import itertools

MATRIX_SIZE = 25
NUM_TRAIN_SAMPLES = 800
NUM_TEST_SAMPLES = 200
BATCH_SIZE_PER_REPLICA = 32

KAGGLE_BASE_DIR = "/kaggle/working/"
BASE_OUTPUT_DIR = KAGGLE_BASE_DIR
BASE_DATASET_DIR = os.path.join(BASE_OUTPUT_DIR, "datasets")
BASE_RESULTS_DIR = os.path.join(BASE_OUTPUT_DIR, "results")

def euclidean_distance(p1, p2):
    return np.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

def generate_problem_a_sample():
    matrix = np.zeros((MATRIX_SIZE, MATRIX_SIZE), dtype=int)
    points_coords = []
    while len(points_coords) < 2:
        r, c = np.random.randint(0, MATRIX_SIZE, 2)
        if (r, c) not in points_coords:
            points_coords.append((r, c))
            matrix[r, c] = 1
    distance = euclidean_distance(points_coords[0], points_coords[1])
    return matrix, distance

def generate_problem_b_sample():
    matrix = np.zeros((MATRIX_SIZE, MATRIX_SIZE), dtype=int)
    num_points = np.random.randint(3, 11)
    points_coords = []
    while len(points_coords) < num_points:
        r, c = np.random.randint(0, MATRIX_SIZE, 2)
        if (r,c) not in points_coords and matrix[r,c] == 0:
            points_coords.append((r,c))
            matrix[r,c] = 1
    
    if len(points_coords) < 2:
        return generate_problem_b_sample()

    min_dist = float('inf')
    for p1, p2 in itertools.combinations(points_coords, 2):
        dist = euclidean_distance(p1, p2)
        if dist < min_dist:
            min_dist = dist
    return matrix, min_dist

def generate_problem_c_sample():
    matrix = np.zeros((MATRIX_SIZE, MATRIX_SIZE), dtype=int)
    num_points = np.random.randint(3, 11)
    points_coords = []
    while len(points_coords) < num_points:
        r, c = np.random.randint(0, MATRIX_SIZE, 2)
        if (r,c) not in points_coords and matrix[r,c] == 0:
            points_coords.append((r,c))
            matrix[r,c] = 1

    if len(points_coords) < 2:
        return generate_problem_c_sample()

    max_dist = 0.0
    for p1, p2 in itertools.combinations(points_coords, 2):
        dist = euclidean_distance(p1, p2)
        if dist > max_dist:
            max_dist = dist
    return matrix, max_dist

def generate_problem_d_sample():
    matrix = np.zeros((MATRIX_SIZE, MATRIX_SIZE), dtype=int)
    num_points_to_place = np.random.randint(1, 11)
    actual_points_placed = 0
    placed_coordinates = set()
    attempts, max_attempts = 0, MATRIX_SIZE * MATRIX_SIZE * 2

    while actual_points_placed < num_points_to_place and attempts < max_attempts:
        r, c = np.random.randint(0, MATRIX_SIZE, 2)
        if (r, c) not in placed_coordinates:
            matrix[r, c] = 1
            placed_coordinates.add((r,c))
            actual_points_placed += 1
        attempts += 1
    return matrix, float(actual_points_placed)

def generate_problem_e_sample():
    matrix = np.zeros((MATRIX_SIZE, MATRIX_SIZE), dtype=int)
    num_squares_to_draw = np.random.randint(1, 11)
    for _ in range(num_squares_to_draw):
        square_size = np.random.randint(1, MATRIX_SIZE + 1)
        top_r = np.random.randint(0, MATRIX_SIZE - square_size + 1)
        top_c = np.random.randint(0, MATRIX_SIZE - square_size + 1)
        matrix[top_r : top_r + square_size, top_c : top_c + square_size] = 1
    return matrix, float(num_squares_to_draw)

def generate_dataset_for_problem_type(problem_func, problem_letter, num_samples, base_dataset_path, dataset_split_name):
    data_x_filepath = os.path.join(base_dataset_path, f"problem_{problem_letter}_{dataset_split_name}_x.npy")
    data_y_filepath = os.path.join(base_dataset_path, f"problem_{problem_letter}_{dataset_split_name}_y.npy")

    if os.path.exists(data_x_filepath) and os.path.exists(data_y_filepath):
        print(f"Dataset {problem_letter} ({dataset_split_name}) already exists at {base_dataset_path}. Skipping generation.")
        return

    print(f"Generating {dataset_split_name} dataset for Problem {problem_letter} ({num_samples} samples)...")
    data_x = []
    data_y = []
    for i in range(num_samples):
        if (i + 1) % 100 == 0:
            print(f"  Generated {i + 1}/{num_samples} samples for Problem {problem_letter} ({dataset_split_name})")
        matrix, output_value = problem_func()
        data_x.append(matrix.reshape(MATRIX_SIZE, MATRIX_SIZE, 1))
        data_y.append(output_value)
    
    np.save(data_x_filepath, np.array(data_x))
    np.save(data_y_filepath, np.array(data_y))
    print(f"Saved {dataset_split_name} dataset for Problem {problem_letter} to {base_dataset_path}")

def run_all_dataset_generation(base_dataset_path):
    os.makedirs(base_dataset_path, exist_ok=True)
    
    problem_generators = {
        'A': generate_problem_a_sample, 'B': generate_problem_b_sample,
        'C': generate_problem_c_sample, 'D': generate_problem_d_sample,
        'E': generate_problem_e_sample
    }
    for p_letter, p_func in problem_generators.items():
        generate_dataset_for_problem_type(p_func, p_letter, NUM_TRAIN_SAMPLES, base_dataset_path, "train")
        generate_dataset_for_problem_type(p_func, p_letter, NUM_TEST_SAMPLES, base_dataset_path, "test")
    print("All dataset generation attempts completed.")

def load_problem_dataset(problem_letter, base_dataset_path):
    print(f"Attempting to load dataset for Problem {problem_letter} from base path: {base_dataset_path}")
    train_x_path = os.path.join(base_dataset_path, f"problem_{problem_letter}_train_x.npy")
    train_y_path = os.path.join(base_dataset_path, f"problem_{problem_letter}_train_y.npy")
    test_x_path = os.path.join(base_dataset_path, f"problem_{problem_letter}_test_x.npy")
    test_y_path = os.path.join(base_dataset_path, f"problem_{problem_letter}_test_y.npy")

    required_files = [train_x_path, train_y_path, test_x_path, test_y_path]
    if not all(os.path.exists(p) for p in required_files):
        print(f"CRITICAL: Dataset files missing for Problem {problem_letter} in {base_dataset_path}. Expected:")
        for p in required_files: print(f"  {p} (Found: {os.path.exists(p)})")
        return None, None, None, None
    try:
        train_x = np.load(train_x_path)
        train_y = np.load(train_y_path)
        test_x = np.load(test_x_path)
        test_y = np.load(test_y_path)
        print(f"Dataset for Problem {problem_letter} loaded successfully. Shapes: train_x={train_x.shape}, test_x={test_x.shape}")
        return train_x, train_y, test_x, test_y
    except Exception as e:
        print(f"Error loading dataset files from {base_dataset_path} for Problem {problem_letter}: {e}")
        return None, None, None, None

def create_cnn_model(input_shape, num_outputs=1, output_activation='linear'):
    model = keras.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape, padding='same'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(num_outputs, activation=output_activation)
    ])
    return model

def main_training_evaluation(base_dataset_path, base_results_path):
    try:
        tpu = tf.distribute.cluster_resolver.TPUClusterResolver.connect()
        strategy = tf.distribute.TPUStrategy(tpu)
        print('Running on TPU:', tpu.master())
        print("Number of TPU replicas:", strategy.num_replicas_in_sync)
    except ValueError:
        print("TPU not found. Falling back to CPU/GPU strategy.")
        strategy = tf.distribute.get_strategy()

    GLOBAL_BATCH_SIZE = BATCH_SIZE_PER_REPLICA * strategy.num_replicas_in_sync
    print(f"Global batch size: {GLOBAL_BATCH_SIZE}")

    problems_to_run = ['A', 'B', 'C', 'D', 'E']
    data_fractions = [0.25, 0.50, 1.00] 
    fraction_labels = ["quarter", "half", "full"]

    for current_problem_choice in problems_to_run:
        print(f"\n\n--- Starting Processing for Problem {current_problem_choice} using CNN model ---")

        problem_results_path_base = os.path.join(base_results_path, f"problem_{current_problem_choice}")
        os.makedirs(problem_results_path_base, exist_ok=True)

        train_x_full, train_y_full, test_x, test_y = load_problem_dataset(current_problem_choice, base_dataset_path=base_dataset_path)
        
        if train_x_full is None or train_y_full is None or test_x is None or test_y is None:
            print(f"Skipping Problem {current_problem_choice} due to data loading failure.")
            continue

        train_x_full_reshaped = None
        test_x_reshaped = None

        if train_x_full.ndim == 4 and train_x_full.shape[1:] == (MATRIX_SIZE, MATRIX_SIZE, 1):
            train_x_full_reshaped = train_x_full
        elif train_x_full.ndim == 3 and train_x_full.shape[1:] == (MATRIX_SIZE, MATRIX_SIZE):
            print(f"Warning: train_x_full for Problem {current_problem_choice} has ndim=3. Expanding dims.")
            train_x_full_reshaped = np.expand_dims(train_x_full, axis=-1)
        else:
            print(f"Error: train_x_full for Problem {current_problem_choice} has unexpected shape {train_x_full.shape}. Skipping.")
            continue
            
        if test_x.ndim == 4 and test_x.shape[1:] == (MATRIX_SIZE, MATRIX_SIZE, 1):
            test_x_reshaped = test_x
        elif test_x.ndim == 3 and test_x.shape[1:] == (MATRIX_SIZE, MATRIX_SIZE):
            print(f"Warning: test_x for Problem {current_problem_choice} has ndim=3. Expanding dims.")
            test_x_reshaped = np.expand_dims(test_x, axis=-1)
        else:
            print(f"Error: test_x for Problem {current_problem_choice} has unexpected shape {test_x.shape}. Skipping.")
            continue

        input_shape_for_model = (MATRIX_SIZE, MATRIX_SIZE, 1)
        
        max_train_val = np.max(train_x_full_reshaped) if np.max(train_x_full_reshaped) > 0 else 1.0
        train_x_full_processed = train_x_full_reshaped.astype('float32') / max_train_val
        
        max_test_val = np.max(test_x_reshaped) if np.max(test_x_reshaped) > 0 else 1.0
        test_x_processed = test_x_reshaped.astype('float32') / max_test_val

        train_y_full_processed = train_y_full.astype('float32')
        test_y_processed = test_y.astype('float32')

        for i_frac, fraction in enumerate(data_fractions):
            label = fraction_labels[i_frac]
            print(f"\n\n--- Training with {label} ({fraction*100}%) of training data for Problem {current_problem_choice} ---")

            num_subset_samples = int(len(train_x_full_processed) * fraction)
            train_x_subset = train_x_full_processed[:num_subset_samples]
            train_y_subset = train_y_full_processed[:num_subset_samples]

            if len(train_x_subset) == 0:
                print(f"Skipping {label} fraction as it results in 0 samples for Problem {current_problem_choice}.")
                continue
            
            print(f"Using {len(train_x_subset)} samples for training ({label} data) for Problem {current_problem_choice}.")

            current_fraction_results_path = os.path.join(problem_results_path_base, label)
            os.makedirs(current_fraction_results_path, exist_ok=True)

            problem_type = 'regression'
            num_outputs = 1
            output_activation = 'linear'
            loss_function = 'mean_squared_error'
            metrics_list = ['mean_absolute_error', 'mean_squared_error']

            with strategy.scope():
                model = create_cnn_model(input_shape_for_model, num_outputs, output_activation)
                optimizer = keras.optimizers.Adam(learning_rate=0.001)
                model.compile(optimizer=optimizer, loss=loss_function, metrics=metrics_list)
            
            if i_frac == 0: 
                 model.summary()

            model_checkpoint_path = os.path.join(current_fraction_results_path, f"best_model_problem_{current_problem_choice}_CNN_{label}.keras")
            callbacks = [
                keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1),
                keras.callbacks.ModelCheckpoint(filepath=model_checkpoint_path, save_best_only=True, monitor='val_loss', verbose=1)
            ]

            print(f"\n--- Training CNN model for Problem {current_problem_choice} with {label} data ---")
            history = model.fit(train_x_subset, train_y_subset,
                                  epochs=50,
                                  batch_size=GLOBAL_BATCH_SIZE,
                                  validation_split=0.2,
                                  callbacks=callbacks,
                                  verbose=1)
            
            print(f"Loading best model from {model_checkpoint_path} for final evaluation (trained on {label} data).")
            model = keras.models.load_model(model_checkpoint_path)

            print(f"\n--- Evaluating CNN model for Problem {current_problem_choice} (trained on {label} data) on the FULL test set ---")
            results = model.evaluate(test_x_processed, test_y_processed, batch_size=GLOBAL_BATCH_SIZE, verbose=1)
            print(f"Test Results for Problem {current_problem_choice} (CNN, trained on {label} data):")
            print(f"  Loss ({loss_function}): {results[0]:.4f}")
            for idx, metric_name in enumerate(model.metrics_names[1:]):
                print(f"  Metric ({metric_name}): {results[idx+1]:.4f}")

            print(f"\n--- Generating Combined Plot for Problem {current_problem_choice} (CNN, trained on {label} data) ---")
            test_predictions = model.predict(test_x_processed, batch_size=GLOBAL_BATCH_SIZE).flatten()
            
            fig_combined, axes_combined = plt.subplots(1, 3, figsize=(24, 6)) # 1 row, 3 columns
            fig_combined.suptitle(f'Problem {current_problem_choice} - CNN Performance ({label} Training Data)', fontsize=18, y=1.02)

            if hasattr(history, 'history') and 'loss' in history.history:
                axes_combined[0].plot(history.history['loss'], label='Training Loss')
                if 'val_loss' in history.history:
                    axes_combined[0].plot(history.history['val_loss'], label='Validation Loss')
                axes_combined[0].set_title('Model Loss')
                axes_combined[0].set_xlabel('Epoch')
                axes_combined[0].set_ylabel('Loss')
                axes_combined[0].legend()
                axes_combined[0].grid(True)
            else:
                axes_combined[0].text(0.5, 0.5, 'No training history data', ha='center', va='center')
                axes_combined[0].set_title('Model Loss')

            axes_combined[1].scatter(test_y_processed, test_predictions, alpha=0.5, edgecolors='k', s=20)
            axes_combined[1].set_xlabel("Actual Values")
            axes_combined[1].set_ylabel("Predicted Values")
            axes_combined[1].set_title('Predicted vs. Actual')
            if len(test_y_processed) > 0 and len(test_predictions) > 0:
                lims = [
                    min(min(test_y_processed), min(test_predictions)),
                    max(max(test_y_processed), max(test_predictions))
                ]
                axes_combined[1].plot(lims, lims, 'k-', alpha=0.75, zorder=0)
            axes_combined[1].grid(True)

            errors = test_predictions - test_y_processed
            axes_combined[2].hist(errors, bins=25, edgecolor='k')
            axes_combined[2].set_xlabel("Prediction Error (Predicted - Actual)")
            axes_combined[2].set_ylabel("Frequency")
            axes_combined[2].set_title('Residuals Distribution')
            axes_combined[2].grid(True)

            plt.tight_layout(rect=[0, 0, 1, 0.97]) # Adjust layout to make space for suptitle
            plot_filename_combined = os.path.join(current_fraction_results_path, f"combined_analysis_CNN_{label}.png")
            plt.savefig(plot_filename_combined)
            print(f"Combined analysis plot saved to {plot_filename_combined}")
            plt.show()
            plt.close(fig_combined)
            
            num_predictions_to_show = 10
            print(f"\nShowing first {num_predictions_to_show} predictions vs actuals for Problem {current_problem_choice}, {label} data:")
            for i_pred in range(min(num_predictions_to_show, len(test_predictions))):
                print(f"  Sample {i_pred+1}: Predicted={test_predictions[i_pred]:.4f}, Actual={test_y_processed[i_pred]:.4f}")
            
            print(f"Best CNN model for Problem {current_problem_choice} (trained on {label} data) saved at {model_checkpoint_path}")
            print(f"--- Finished Processing for Problem {current_problem_choice} with {label} data ---")
        
        print(f"--- Finished All Fraction Processing for Problem {current_problem_choice} ---")

    print("\n\nAll problems and fractions processed.")

if __name__ == '__main__':
    print(f"TensorFlow Version: {tf.__version__}")
    print(f"Keras Version: {keras.__version__}")
    print(f"Using base output directory: {BASE_OUTPUT_DIR}")
    print(f"Datasets will be stored in/loaded from: {BASE_DATASET_DIR}")
    print(f"Results (models, plots) will be stored in: {BASE_RESULTS_DIR}")

    os.makedirs(BASE_DATASET_DIR, exist_ok=True)
    os.makedirs(BASE_RESULTS_DIR, exist_ok=True)

    run_all_dataset_generation(BASE_DATASET_DIR)
    
    main_training_evaluation(BASE_DATASET_DIR, BASE_RESULTS_DIR)
```
